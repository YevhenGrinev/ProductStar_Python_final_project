```python
df1 = df.groupby('date', as_index=False).sum()[['total']].plot()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['isoYear'] = df['date'].dt.isocalendar().year 
df['week'] = df['date'].dt.isocalendar().week 
df['weekday'] = df['date'].dt.weekday
df['startOfWeek'] = df['date'] - df['date'].dt.weekday * pd.Timedelta(days=1) 
df['day'] = df['date'].dt.dayofyear


df.isoYear.unique()
# df.dtypes
# df[17500:17550]
```

```python
df_t = df.groupby(['year', 'day'],as_index=False).sum() 
g = sns.relplot(x='day', y='total', hue='year', kind='line', data=df_t)

df_t = df.groupby(['startOfWeek'],as_index=False).sum() 
g = sns.relplot(x='startOfWeek', y='total', kind='line', data=df_t) 
g.fig.autofmt_xdate()

df_t = df.groupby(['year','week'],as_index=False).sum() 
g = sns.relplot(x='week', y='total', hue='year', kind='line', data=df_t)
```

Задание 3.1. Сколько групп с разной сезонностью можно выделить в товарах?

Нужно выделить различную сезонность товаров. Для этого, например, можно посчитать долю выручки в квартал для каждого item. И построить relplot попарно по этим долям, чтобы понять, как можно эти группы выделить. Для удобства берём 1 год.

```python
# df_2018 = df[df['year'] == 2018]
# # df_2018.tail()
# df_2018.loc[:, 'quarter'] = 1
# df_2018.loc[df_2018['date'].dt.month >= 4, 'quarter'] = 2
# df_2018.loc[df_2018['date'].dt.month >= 7, 'quarter'] = 3
# df_2018.loc[df_2018['date'].dt.month >= 10, 'quarter'] = 4

# total = df.total.sum()
# total

# df_2018_quarters_items = df_2018.groupby(['quarter', 'item'], as_index=False).sum()[['quarter', 'item', 'total']]
# df_2018_quarters_items

# df_2018.loc[df_2018['quarter'].dt.month == 1, 'proportion'] = df_2018.total / df_2018_quarters_items
# df_2018.loc[df_2018['quarter'].dt.month == 2, 'proportion'] = df_2018.total / df_2018_quarters_items
# df_2018.loc[df_2018['quarter'].dt.month == 3, 'proportion'] = df_2018.total / df_2018_quarters_item
# df_2018.loc[df_2018['quarter'].dt.month == 4, 'proportion'] = df_2018.total / df_2018_quarters_items
# df_2018.head()
```

```python
df_t = df[df['date'].dt.year==2018].copy()
df_t.loc[:,'qrt'] = 0
df_t.loc[df_t['date'].dt.month >= 4,'qrt'] = 1
df_t.loc[df_t['date'].dt.month >= 7,'qrt'] = 2
df_t.loc[df_t['date'].dt.month >= 10,'qrt'] = 3
# df_t

df_t = df_t[['item','qrt','total']].groupby(['item','qrt']).sum().unstack()
# df_t
df_t.columns = [0, 1, 2, 3]
# df_t
df_t['s'] = df_t[0] + df_t[1] + df_t[2] + df_t[3]
df_t[0] = df_t[0]/df_t['s']
df_t[1] = df_t[1]/df_t['s']
df_t[2] = df_t[2]/df_t['s']
df_t[3] = df_t[3]/df_t['s']
# df_t
sns.relplot(data=df_t, x=0, y=1)
sns.relplot(data=df_t, x=2, y=3)
sns.relplot(data=df_t, x=0, y=2)
```

Задание 3.2. Какое соотношение продаж по ним?

Сначала нужно itemам приписать группы.

```python
df_t.loc[:, 'tp'] = 'I'
df_t.loc[df_t[2] > 0.29, 'tp'] = 'II'
df_t.loc[df_t[3] > 0.37, 'tp'] == 'III'
df_m = pd.DataFrame(df_t['tp'], index=df_t.index).reset_index()
# df_m
df_tt = df.merge(df_m, on='item').groupby(['tp', 'startOfWeek'], as_index=False).sum() 
g = sns.relplot(data=df_tt, x='startOfWeek', y='total', hue='tp', kind='line')
g.fig.autofmt_xdate()
```
